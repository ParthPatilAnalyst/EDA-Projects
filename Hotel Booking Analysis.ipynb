```python
df['country']=df['country'].fillna(df['country'].mode()[0])
```

```python
df['agent'] = df['agent'].fillna(0).astype(int)
df['company'] = df['company'].fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])


month_map = {m:i for i,m in enumerate(['January','February','March','April','May','June','July','August','September','October','November','December'],1)}
df['arrival_date_month_num'] = df['arrival_date_month'].map(month_map)


df['arrival_date'] = pd.to_datetime(dict(year=df['arrival_date_year'],
                                         month=df['arrival_date_month_num'],
                                         day=df['arrival_date_day_of_month']), errors='coerce')


df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
```

```python
df.drop_duplicates(inplace = True)
```

# Exploratory Data Analysis (EDA)

## Basic Level

#### What is the average lead time for bookings?

```python
a = df['lead_time'].mean()
a
```

#### What is the distribution of bookings by hotel type?

```python
ax = df['hotel'].value_counts().plot(kind='bar')

for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2, 
            i.get_height(), 
            int(i.get_height()), 
            ha='center')

plt.show()
```

#### How many bookings were canceled?

```python
ax = df['is_canceled'].value_counts().plot(kind='bar')

for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2, 
            i.get_height(), 
            int(i.get_height()), 
            ha='center')

plt.show() 

#1 = canceled, 0 = not canceled.
```

#### What is the most common arrival month for bookings?

```python
most_common_month = df['arrival_date_month'].mode()[0]
month_counts = df['arrival_date_month'].value_counts()

plt.figure(figsize=(12,6))   

ax = month_counts.plot(kind='bar', title='Bookings by Arrival Month')


for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2,
            i.get_height(),
            int(i.get_height()),
            ha='center',
            va='bottom')

plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

most_common_month, month_counts
```

#### What is the average number of special requests per booking?

```python
df['total_of_special_requests'].mean()
```

#### Which country has the highest number of bookings?

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(12,6))

ax = top_countries.plot(kind='bar', title='Top 10 Countries by Bookings')

for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2,
            i.get_height(),
            int(i.get_height()),
            ha='center',
            va='bottom')

plt.ylabel('Bookings')
plt.xticks(rotation=45)
plt.show()
```

#### What is the average daily rate (ADR) for each hotel type?

```python
df.groupby('hotel')['adr'].mean().round(2)
```

#### What percentage of guests required car parking spaces?

```python
(df['required_car_parking_spaces']>0).mean()*100
```

#### What is the average stay duration in week nights and weekend nights?

```python
print(f'the average stay duration in week nights is {df['stays_in_week_nights'].mean().round(2)}')
print(f'the average stay duration in  weekend nights is {df['stays_in_weekend_nights'].mean().round(2)}')
```

#### How many bookings were made through travel agents?

```python
bookings_by_agents = (df['agent'] != 0).sum()
print(f'{((bookings_by_agents / len(df)) * 100).round(2)}%')
```

### Medium Level

#### What is the cancellation rate for each hotel type?

```python
ax = (df.groupby('hotel')['is_canceled'].mean() * 100).plot(kind = 'bar')
for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2, 
            i.get_height(), 
            int(i.get_height()), 
            ha='center')

plt.show() 
```

#### What is the average ADR per market segment?

```python
df.groupby('market_segment')['adr'].mean()
```

#### What is the relationship between lead time and cancellation rate?

```python
df['lead_time_bin'] = pd.cut(df['lead_time'], bins=10)
cancel_rate = df.groupby('lead_time_bin')['is_canceled'].mean()
plt.figure(figsize=(10,6))
cancel_rate.plot(kind='line')

plt.ylabel('Cancellation Rate')
plt.title('Lead Time vs Cancellation Rate')
plt.xticks(rotation=45)
plt.show()





plt.figure(figsize=(10,6))

plt.scatter(df['lead_time'], df['is_canceled'], alpha=0.3)

plt.xlabel('Lead Time')
plt.ylabel('Is Canceled (0 = No, 1 = Yes)')
plt.title('Lead Time vs Cancellation')

plt.show()
```

#### Which distribution channel has the highest number of bookings?

```python
df['distribution_channel'].value_counts().plot(kind='bar', title='Distribution Channel Bookings')
plt.ylabel('Count')
plt.show()
```

#### What is the average number of previous cancellations by hotel type?

```python
df.groupby('hotel')['previous_cancellations'].mean()
```

#### What is the trend of ADR over the years?

```python
plt.figure(figsize=(10,6))
df.groupby('arrival_date_year')['adr'].mean().plot(marker='o', title='Average ADR by Year')
plt.ylabel('ADR')
plt.show()
```

#### Which month has the highest revenue?

```python
df['revenue'] = df['adr'] * df['total_nights']
plt.figure(figsize=(10,6))
monthly_revenue = df.groupby('arrival_date_month')['revenue'].sum().reindex([
    'January','February','March','April','May','June','July','August','September','October','November','December'
]).plot(kind='bar', title='Total Revenue by Month')
plt.ylabel('Revenue (estimated)')
plt.show()
```

#### What is the impact of special requests on ADR?

```python
plt.figure(figsize=(10,6))
df.groupby('total_of_special_requests')['adr'].mean().plot(marker='o', title='ADR by Number of Special Requests')
plt.ylabel('ADR')
plt.show()
```

#### Average stay duration for repeated guests vs new guests?

```python
df.groupby('is_repeated_guest')['total_nights'].mean().rename(index={0:'new',1:'repeated'})
```

#### Which room type has the highest number of bookings?

```python
top_rooms = df['reserved_room_type'].value_counts().head(5)
plt.figure(figsize=(10,6))
ax = top_rooms.plot(kind='bar', title='Top 5 Reserved Room Types')
for i in ax.patches:
    ax.text(i.get_x() + i.get_width()/2,
            i.get_height(),
            int(i.get_height()),
            ha='center',
            va='bottom')
plt.ylabel('Bookings')
plt.show()
```

```python
df['Rating']=df.groupby('Category')['Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
df.drop_duplicates(inplace = True)
```

```python
df['Reviews'] = df['Reviews'].astype('int')
```

```python
df['Size'].unique()
```

```python
df['Size']=df['Size'].str.replace('M','000')
df['Size']=df['Size'].str.replace('k','')
df['Size']=df['Size'].replace('Varies with device',np.nan)
```

```python
df['Size'] = df['Size'].astype('float')
```

```python
df['Size'] = df['Size'].fillna(df['Size'].median())
```

```python
df.info()
df
```

```python
df['Installs'] = df['Installs'].str.replace('+','')
df['Installs'] = df['Installs'].str.replace(',','')
df['Installs'] = df['Installs'].astype('int')
```

```python
df.info()
```

```python
df['Price'].unique()
```

```python
df['Price'] = df['Price'].str.replace('$','')
```

```python
df['Price'] = df['Price'].astype('float')
```

```python
df['Last Updated'] = pd.to_datetime(df['Last Updated'])
```

```python
df.info()
```

```python
df.duplicated(subset = ['App']).sum()
```

```python
df.drop_duplicates(subset=['App'],inplace = True)
```

#### Analyze the trend of app updates over time. Are there any noticeable patterns or seasonal trends?

```python
df['Year'] = df['Last Updated'].dt.year
df['Month'] = df['Last Updated'].dt.month

updates_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=updates_per_year.index, y=updates_per_year.values, marker='o')
plt.title("Number of App Updates per Year")
plt.xlabel("Year")
plt.ylabel("Number of Updates")
plt.show()

updates_per_month = df['Month'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=updates_per_month.index, y=updates_per_month.values, marker='o')
plt.title("Seasonal Trend: App Updates by Month (All Years)")
plt.xlabel("Month")
plt.ylabel("Number of Updates")
plt.xticks(range(1,13))
plt.show()
```

#### How does the average rating of apps change with the number of installs? Create a binned analysis.

```python
bins = [0, 10000, 100000, 1000000, 10000000, 100000000]
labels = ['0-10K', '10K-100K', '100K-1M', '1M-10M', '10M+']

df['Install_Bin'] = pd.cut(df['Installs'], bins=bins, labels=labels)
avg_rating = df.groupby('Install_Bin')['Rating'].mean()
avg_rating
avg_rating.plot(kind='bar')
plt.title("Average Rating by Install Range")
plt.ylabel("Average Rating")
plt.show()
```

#### What is the relationship between app genre and user ratings? Are certain genres consistently rated higher or lower

```python
genre_rating = df.groupby('Genres')['Rating'].agg(['mean', 'median']).sort_values(by='mean', ascending=False)

genre_rating
genre_rating['mean'].sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Genres by Average Rating")
plt.ylabel("Average Rating")
plt.show()
```
